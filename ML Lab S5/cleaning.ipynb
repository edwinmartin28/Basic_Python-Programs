```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as pl
dataset=pd.read_csv('hello.csv')
dataset.head(200)


```

```python
dataset.isnull().sum() #number of null values column wise
```

```python
dataset=dataset.dropna() #deletes all rows containing null values
```

```python
dataset['AUD/JPY (open)'].mean() #for converting null values to numerical values
```

```python
dataset.head(5)
```

```python
dataset['AUD/JPY (open)']=dataset['AUD/JPY (open)'].replace(np.NaN,dataset['AUD/JPY (open)'].mean()).head(160)
dataset.head(5)
```

```python
dataset['AUD/JPY (high)']=dataset['AUD/JPY (high)'].replace(np.NaN,dataset['AUD/JPY (high)'].mean()).head(160)
dataset.head(5)
```
